#### 1) Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um método para calcular a média e outro para verificar se o aluno foi aprovado (média >= 6).

```python
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas

    def calcular(self):
        return sum(self.notas) / len(self.notas) 

    def aprovado(self):
        media = self.calcular()
        if media >= 6:
            return f"Aluno {self.nome} aprovado com média: {media:.2f}"
        else:
            return f"Aluno {self.nome} reprovado com média: {media:.2f}"
        
aluno1 = Aluno("João", [10, 9, 3])
aluno2 = Aluno("Pedro", [1, 9, 3])
print(aluno1.aprovado())
print(aluno2.aprovado())

    
```

#### 2) Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente métodos para atualizar a taxa (com validação para não aceitar valores negativos) e para converter um valor de Real para Dólar.

```python
class Conversor:
    def __init__(self, taxa_dolar):
        self.__dolar = taxa_dolar

    def atualizar_taxa(self, nova_taxa):
        if nova_taxa < 0:
            return f"A taxa não pode ser negativa"
        else:
            self.__dolar = nova_taxa
            return f"A nova taxa é: {self.__dolar:.2f}"

    def converter(self, valor_real):
        return f"O valor da conversão de R$ {valor_real} é de: $ {valor_real / self.__dolar:.2f}"
    
conversor = Conversor(0.19)
print(conversor.converter(1))

print(conversor.atualizar_taxa(-1))
print(conversor.atualizar_taxa(0.50))
print(conversor.converter(1))
```

#### 3) Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os títulos disponíveis.

```python
class Livro:
    def __init__(self, titulos):
        self.titulos = titulos

class Biblioteca():
    def __init__(self):
        self.livros = []

    def adicionar_livros(self, livro):
        self.livros.append(livro)

    def listar(self):
        return f"A lista de títulos disponíveis são: \n" + "\n".join(livro.titulos for livro in self.livros)
    

livro = Livro("Diário de um banana")
livro2 = Livro("João e Maria")

biblioteca = Biblioteca()
biblioteca.adicionar_livros(livro)
biblioteca.adicionar_livros(livro2)

print(biblioteca.listar())
    
```

#### 4) Crie uma classe base Veiculo com atributos como marca e modelo. Crie subclasses Carro (com número de portas) e Moto (com cilindradas). Ambas devem herdar as características da classe pai.

```python
class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def listar(self):
        return f"marca {self.marca} modelo {self.modelo}"
    
class Carro(Veiculo):
    def __init__(self, marca, modelo, nro_portas):
        super().__init__(marca, modelo)
        self.nro_portas = nro_portas

    def listar_carro(self):
        return f"O carro é da {super().listar()} e possui {self.nro_portas} portas"
    
class Moto(Veiculo):
    def __init__(self, marca, modelo, cilindradas):
        super().__init__(marca, modelo)
        self.cilindradas = cilindradas

    def listar_moto(self):
        return f"A moto é da {super().listar()} e possui {self.cilindradas} cilindradas"

veiculo1 = Veiculo("Chevrolet", "Onix")
veiculo2 = Veiculo("Honda", "CB500F")
carro1 = Carro(veiculo1.marca, veiculo1.modelo, 4)
moto1 = Moto(veiculo2.marca, veiculo2.modelo, 500)

print(carro1.listar_carro())
print(moto1.listar_moto())
```

#### 6) No contexto de IoT, crie uma classe Sensor que armazene ID e tipo. Crie um script que permita ao usuário cadastrar vários sensores em um dicionário, onde a chave é o ID e o valor é o objeto Sensor.

```python
class Sensor:
    def __init__(self, ID, tipo):
        self.id = ID
        self.tipo = tipo

sensores = {}

while True:
    ID = input("Digite o ID do sensor (ou 'sair' para encerrar): ")
    if ID.lower() == 'sair':
        break
    if ID in sensores:
        print("ID já cadastrado. Tente outro.")
        continue
    tipo = input("Digite o tipo do sensor: ")

    sensores[ID] = {"id": ID, "tipo": tipo}
    print(f"Sensor {ID} do tipo {tipo} cadastrado com sucesso!")

print(sensores)
```

#### 7) Crie uma classe Usuario com login e senha. Implemente um método que valide se a senha possui pelo menos 8 caracteres e um caractere especial. Se não atender, o sistema deve solicitar uma nova senha.

```python
import re

class Usuario:
    def __init__(self, login, senha):
        self.login = login
        self.senha = senha

    def validacao(self):
        caracteres_especiais = r'[!@#$%^&*(),.?":{}|<>]'
        while True:
            if len(self.senha) >=8 and re.search(caracteres_especiais, self.senha):
                print("Senha válida")
                break
            else:
                print("Senhá inválida, precisa ter pelo menos 8 caracteres e 1 sendo especial") 
                self.senha = input("Digite uma nova senha:")
    
usuario1 = Usuario("ana", "123")
usuario1.validacao()

print(f"O login é: {usuario1.login} e a senha é: {usuario1.senha}")
        
```

#### 8) Crie uma classe Camera e uma classe Telefone. Crie uma terceira classe Smartphone que herde de ambas, demonstrando como o objeto final pode acessar métodos de as duas classes originais.

```python
class Camera:
    def tipo(self):
        return f"Câmera"

class Telefone:
    def tipo2(self):
        return f"Telefone"

class Smartphone(Camera, Telefone):
    def info(self):
        return f"Método da classe Câmera: {self.tipo()} - Método da classe Telefone: {self.tipo2()}"
    
smartphone = Smartphone()

print(smartphone.tipo())
print(smartphone.tipo2())
print(smartphone.info())
```

#### 9) Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o módulo datetime para capturar o horário exato em que o método foi chamado e armazene isso em uma lista de logs dentro do objeto.

```python
import datetime

class Funcionario:
    def __init__(self):
        self.logs = []

    def registrar_entrada(self):
        hora = datetime.datetime.now().time()
        self.logs.append(hora) 
        
        return f"Logs de entrada: \n" + "\n".join(h.strftime('%H:%M:%S') for h in self.logs)

funcionario = Funcionario()
print(funcionario.registrar_entrada())
print(funcionario.registrar_entrada())
```

### 10) Crie uma classe Calculadora que contenha apenas métodos estáticos para operações como soma, subtração e raiz quadrada. O objetivo é usar a classe como um "namespace" de funções utilitárias sem precisar instanciá-la.

```python
class Calculadora:
    @staticmethod
    def soma(*numeros):
        return sum(numeros)
    
    @staticmethod
    def subtracao(*numeros):
        resultado = numeros[0]
        for n in numeros[1:]:
            resultado -= n
        return resultado
     
    @staticmethod
    def raiz_quadrada(a):
        return a ** 0.5

print(Calculadora.soma(9, 3, 2, 1, 23, 4))
print(Calculadora.subtracao(9, 1, 4, 3, 2))
print(Calculadora.raiz_quadrada(9))
```
